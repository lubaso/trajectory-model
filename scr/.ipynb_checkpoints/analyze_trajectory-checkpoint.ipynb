# Analyze trajectories 

## Import libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## Read in data

```python
df_trajectory = pd.read_csv('../output/output_siouxFalls_trajectory.zip')
df_legs = pd.read_csv('../input/output_siouxFalls_legs.gz', sep = ';') 
df_persons = pd.read_csv('../input/output_siouxFalls_persons.gz', sep = ';')  
df_trips = pd.read_csv('../input/output_siouxFalls_trips.gz', sep = ';') 

display(df_trajectory.head())
display(df_legs.head())
display(df_persons.head())
display(df_trips.head())
```

# Visualize data

## Bar plot - Trip mode

```python
#Group by leg; count person id
trips_by_mode = df_trips.groupby('longest_distance_mode')['person'].count()

# Get proportion for each type
trips_pct_by_mode = trips_by_mode / sum(trips_by_mode)
data= pd.DataFrame(data= {'Mode': trips_pct_by_mode.index, 'Percentage': trips_pct_by_mode.values})
print(data)

# Bar plot for trip mode
sns.catplot(x = 'Mode', y = 'Percentage', data = data, kind = "bar")
```

## Histogram - Travel Time

```python
#convert travel time to minutes (int)
def get_sec(time_str):
    """Get minutes from time."""
    h, m, s = time_str.split(':')
    time = int(h) * 60 + int(m) + int(s)/60
    time = round(time)
    return time

df_trips['trav_time'] = df_trips['trav_time'].apply(get_sec)

sns.set_theme()
g = sns.displot(data = df_trips, x = "trav_time", hue = "longest_distance_mode", binwidth = 5, multiple = "stack")
g.set(xlim=(0, 90))
g.set(title = "Histogram - Travel Times",
      xlabel = "Travel time (min)")
g._legend.set_title("Main Mode")
```
